```python
X=df.drop(['HeartDisease'],axis=1)
X.head()
```

```python
y=df['HeartDisease']
y.head()
```

```python
Sex=pd.get_dummies(df['Sex'],drop_first=True)
Sex
```

```python
ChestPainType=pd.get_dummies(df['ChestPainType'],drop_first=True)
ChestPainType
```

```python
RestingECG=pd.get_dummies(df['RestingECG'],drop_first=True)
RestingECG
```

```python
ExerciseAngina=pd.get_dummies(df['ExerciseAngina'],drop_first=True)
ExerciseAngina
```

```python
ST_Slope=pd.get_dummies(df['ST_Slope'],drop_first=True)
ST_Slope
```

```python
X.drop(['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope'],axis=1,inplace=True)
```

```python
X=pd.concat([X,Sex,ChestPainType,RestingECG,ExerciseAngina,ST_Slope],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred=rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
